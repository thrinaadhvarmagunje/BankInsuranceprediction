```python
import numpy as np
```

```python
import pandas as pd
```

```python
df=pd.read_csv("train.csv")
```

```python
df=pd.DataFrame(df)
df=df.loc[:,["age_in_days","Income","target"]]
```

```python

df.dropna(inplace=True)
```

```python
df
```

```python
X=df.iloc[:,:-1]
```

```python
X
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac=accuracy_score(y_test,y_pred)
```

```python
ac
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred1=knn.predict(X_test)
```

```python
y_pred1
```

```python
knnac=accuracy_score(y_test,y_pred1)
```

```python
knnac
```

```python
from sklearn.svm import SVC
```

```python
svc_diag=SVC(C=10,kernel='rbf',gamma=2)
```

```python
svc_diag.fit(X_train,y_train)
```

```python
y_pred2=svc_diag.predict(X_test)
```

```python
y_pred2
```

```python
svmac=accuracy_score(y_test,y_pred1)
```

```python
svmac
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree = dtree.fit(X_train, y_train)
```

```python
ypred3=dtree.predict(X_test)
ypred3

```

```python
dtac=accuracy_score(y_test,ypred3)
```

```python
dtac
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred4= rf.predict(X_test)
```

```python
rfac=accuracy_score(y_test,y_pred4)
rfac
```
